```python
#determine the indexes for the game: I will add 3 times 3 to simulate 3 rows of 3
#This is an example grid

print("Let's play TicTacToe!\n")
grid_ex ='\n0|1|2\n3|4|5\n6|7|8'
print('Positions on the grid:',grid_ex)
          

game_grid = ['*','*','*',
             '*','*','*',
             '*','*','*']
#if game is still going:
game = True

#who wins
winner = None

#turns:
current_player ='X'

# signs =['X','O']

#functions that define what happens in the game

    #This is to have cpu player random (to play against the computer)
    # Idea: to add option 1 or 2 players
    
    # def player_sign (signs):
    #     s=1 
    #     while s==1: 
    #         cpuSign =''
    #         playerSign = (input("\nChoose your sign: ").upper())
    #         if playerSign in signs: 
    #             s=0
    #         else: 
    #             print("\nYour reply is invalid, please try again\n") 
    #         if playerSign == 'X':
    #             cpuSign == 'O'
    #         else:
    #             cpuSign == 'X'
    #     return playerSign, cpuSign

def game_outlook(game_grid):
    
    #this function defines the outlook of the grid and assigns indexes to the positions
    print("\n",game_grid[0]+"|",game_grid[1]+"|",game_grid[2],
          "\n",game_grid[3]+"|",game_grid[4]+"|",game_grid[5],
          "\n",game_grid[6]+"|",game_grid[7]+"|",game_grid[8])

def game_begin():
    #This fuctions defines the begining of the game by calling the game_grid
    game_outlook(game_grid)
    #player_sign(signs)
    while game : 
        game_turn(current_player)
        #we check if there is a winner yet 
        check_if_game_still_on()
        change_player()  
    #the game ends when  
    if winner == "X" or winner == "O":
        print(winner + " is the winner")
    elif winner == None:
        print("Is a tie!")
        
def game_turn(player):
    player_choice = input("choose a spot on the grid :")
        #the type of user_choice is a string bc of the input function so:
    player_choice = int(player_choice)
        
        #now that we have the position = index on the grid so we can place one of the variable 
    game_grid[player_choice] = player
    game_outlook(game_grid)
    


def check_if_game_still_on():
    check_win()
    check_tie()

    
def check_win(): 
    global winner 
    #check if 3 of a kind in Rows (3 rows)
    row_win = check_rows()
    #check if 3 of a kind in columns (3 columns)
    column_win = check_columns()
    #check if 3 of a kind in diagonals (2 diagonals)
    diagonal_win = check_diagonals()
    if row_win:
        #there is a winner
        winner = row_win
    elif column_win:
        #there is a winner
        winner = column_win
    elif diagonal_win:
        #there is a winner
        winner = diagonal_win
    else:
        winner = None
    return

def check_tie():
    global game
    if "*" not in game_grid:
        game=False 
    return



def check_rows():
    #this function determines if we continue to play or not so global variable needs to be called
    global game
    #We have a winning row (r_win)if we have 3 of a kind in our game_grid
    row1_win = game_grid[0]== game_grid[1] == game_grid[2] != '*'
    row2_win = game_grid[3]== game_grid[4] == game_grid[5] != '*'
    row3_win = game_grid[6]== game_grid[7] == game_grid[8] != '*'
    
    #if we have a winning row, the game is set to 0 and the while loop stops
    #if game=0 then we return the value at begining of the row ('X' or 'O')
    if row1_win or row2_win or row3_win:
        game = False
    if row1_win:
        return game_grid[0]
    elif row2_win:
        return game_grid[3]
    elif row3_win:
        return game_grid[6]
    return

def check_columns():
     #this function determines if we continue to play or not so global variable needs to be called
    global game
    #We have a winning column (c_win)if we have 3 of a kind in our game_grid
    col1_win = game_grid[0]== game_grid[3] == game_grid[6] != '*'
    col2_win = game_grid[1]== game_grid[4] == game_grid[7] != '*'
    col3_win = game_grid[2]== game_grid[5] == game_grid[8] != '*'
    
    #if we have a winning column, the game is set to 0 and the while loop stops
    #if game=0 then we return the value at begining of the column ('X' or 'O')
    if col1_win or col2_win or col3_win:
        game = False
    if col1_win:
        return game_grid[0]
    elif col2_win:
        return game_grid[1]
    elif col3_win:
        return game_grid[2]
    return

def check_diagonals():
      #this function determines if we continue to play or not so global variable needs to be called
    global game
    #We have a winning diagonal (d_win)if we have 3 of a kind in our game_grid
    diag1_win = game_grid[0]== game_grid[4] == game_grid[8] != '*'
    diag2_win = game_grid[6]== game_grid[4] == game_grid[2] != '*'
    
    #if we have a winning diagonal, the game is set to 0 and the while loop stops
    #if game=0 then we return the value at begining of the column ('X' or 'O')
    if diag1_win or diag2_win :
        game = False
    if diag1_win:
        return game_grid[8]
    elif diag2_win:
        return game_grid[2]
    return
def change_player():
    #This function will help alternate the players between X and O
    # this first statement is to show that they are equal to, so if it WAS X it becomes O
    global current_player
    if current_player == 'X':
        #this second one is to set the other player to (!= 'equal to')
        current_player = 'O'
    else :
        current_player = 'X'
    return

game_begin()
```
