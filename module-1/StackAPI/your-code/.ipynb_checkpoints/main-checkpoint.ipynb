#### Question 1: Find the questions and answers of last month.

```python
# your code here
SITE = StackAPI('stackoverflow')
questions = SITE.fetch('questions', fromdate=dt.datetime(2020,9,1), todate=dt.datetime(2020,9,30))
questions    
```

#### Question 2: Find the most voted question today with at least a score of 5 and tagged with 'python'. 

```python
# your code here
SITE = StackAPI('stackoverflow')
# min = 4 does not give any results but min=4 gives the result(to show the query is running correctly)
questions_score5 = SITE.fetch('questions', fromdate=1603846810, todate=1603884392, min=4, tagged='python', sort='votes')
questions_score5
```
