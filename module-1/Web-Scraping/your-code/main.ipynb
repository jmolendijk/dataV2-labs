####  Display the 20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe.

````python
# This is the url you will scrape in this exercise
url = 'https://www.emsc-csem.org/Earthquake/'
html = requests.get(url).content
soup = BeautifulSoup(html, "lxml")
````

````python
# Collect and process all latitudes and longitudes to separate list.
cdict = {'class':'tabev1'}
a = soup.find_all('td', cdict)
lst = [a[i].contents[0].replace('\xa0', '') for i in range(len(a))]
lst = lst[:40] #20*2 because includes both latitude and longigude
````

````python
# Collect and process all cardinal directions to separate list.
cdict = {'class':['tabev2']}
a = soup.find_all('td', cdict)    
my_list = [a[i].contents[0].replace('\xa0', '') for i in range(len(a))]
new_list = [x for x in my_list if not any(c.isdigit() for c in x)]
new_list = new_list[:40] #20*2 because includes both latitude and longigude
````

````python
latlong = [(lst[i]+new_list[i]) for i in range(len(lst))]
latlong
latitude = []
longitude = []
for i in range(len(latlong)):
    if i % 2 == 0:
        latitude.append(latlong[i])
    else:
        longitude.append(latlong[i])
print(latitude)
print(longitude)
````

````python
# Collect and process all dates to a separate list.
cdict = {'class':'tabev6'}
dates = soup.find_all('td', cdict)
dates_list = [dates[i].contents[0].contents[1].contents[0] for i in range(len(dates))]
dates_list = [i.replace('\xa0', ' ') for i in dates_list]
dates_list = dates_list[:20]
````

````python
# Collect and process all regions to a separate list.
cdict = {'class':'tb_region'}
regions = soup.find_all('td', cdict)
regions_list = [regions[i].contents[0].replace('\xa0', '') for i in range(len(regions))]
regions_list = regions_list[:20]
````

````python
# Combine all lists into one dataframe.
table = pd.DataFrame(
    {'date-time': dates_list,
     'latitude': latitude,
     'longitude': longitude,
     'regions': regions_list
    })
table
````

#### List all language names and number of related articles in the order they appear in wikipedia.org.

````python
# This is the url you will scrape in this exercise
url = 'https://www.wikipedia.org/'
html = requests.get(url).content
soup = BeautifulSoup(html, "lxml")
````

````python
# your code here
cdict = {'class':'langlist'}
langs = soup.find_all('div', cdict)
````

````python
# Extracting content of all language related tags into one list.
mylist = [langs[i].find_all('a') for i in range(len(langs))]
mylist2 = []
for i in mylist:
    for j in i:
        mylist2.append(j.contents[0])
mylist2
````

````python
# As some elements extracted are tags and others are strings let's split them into two lists and process separatelly.
tags = []
strings = []
for i in mylist2:
    if str(type(i)) == "<class 'bs4.element.Tag'>":
        tags.append(i)
    else:
        strings.append(i)
````

````python
languages = [tags[i].contents[0] for i in range(len(tags))]+strings
languages
````
