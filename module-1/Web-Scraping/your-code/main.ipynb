#### Display the trending Python repositories in GitHub.

The steps to solve this problem is similar to the previous one except that you need to find out the repository names instead of developer names.

````python
# This is the url you will scrape in this exercise
url = 'https://github.com/trending/python?since=daily'
````

````python
html = requests.get(url).content
soup = BeautifulSoup(html, 'html')

text = [element.text for element in soup.find_all('h1',{'class':'h3 lh-condensed'})]
text
#html tag : h1  class="h3 lh-condensed"

repos = []
for repo in text:
    repo = repo.replace('\n', '')
    repo = repo.replace(' ', '')
    repo = repo.strip()
    repos.append(repo)

repos
````

#### Find the number of titles that have changed in the United States Code since its last release point.

````python
# This is the url you will scrape in this exercise
url = 'http://uscode.house.gov/download/download.shtml'
````

````python
html = requests.get(url).content
html
#Titles in bold have been changed since the last release point.

soup = BeautifulSoup(html, 'html')
soup

titles = [element.text for element in soup.find_all('div',{'class':'usctitlechanged'})]
titles

result = []
for title in titles:
    title = title.replace('\n', '')
    title = title.replace('Ù­', '') #The symbol is not * ! Had to copy-paste from the output
    title = title.strip()
    result.append(title)

print(result)

print(len(result), 'titles had been changed since the last release point.')
````

#### List all language names and number of related articles in the order they appear in wikipedia.org.

````python
# This is the url you will scrape in this exercise
url = 'https://www.wikipedia.org/'
````

````python
html = requests.get(url).content
html

soup = BeautifulSoup(html, 'html')
soup

languages = [element.text for element in soup.find_all('strong')]
languages = languages[1:10]
languages

articles = [element.text for element in soup.find_all('small')]
articles
lst_articles = []
for article in articles:
    article = article.replace('\xa0', ',')
    lst_articles.append(article)

lst_articles = lst_articles[:9]
lst_articles

response = {}
for i in range(9):
    response[languages[i]] = lst_articles[i]
    
response
````

#### Display the top 10 languages by number of native speakers stored in a pandas dataframe.

````python
# This is the url you will scrape in this exercise
url = 'https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers'
````

````python
html = requests.get(url).content
soup = BeautifulSoup(html, 'lxml')
soup

table = soup.find_all('table',{'class':'wikitable sortable'})[0]
type(table)

rows = table.find_all('tr')
rows = [row.text.strip().split("\n") for row in rows]
rows

df = pd.DataFrame(rows[1:], columns = rows[0])
df.head(10)
````
