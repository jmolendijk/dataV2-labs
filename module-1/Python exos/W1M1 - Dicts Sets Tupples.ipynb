```python
#Dicts
contacts = {'John': '312-555-1234', 'Paul': '312-555-3123', 'George': '312-555-3333', 'Ringo': '312-555-2222'}
list_contact = print(contacts.keys())
print(contacts.values())
```

```python
print(contacts.items())
```

```python
contacts['Jimmie'] = '312-555-1111'
```

```python
print(contacts.items())
```

```python
del contacts['Jimmie']
```

```python
for i in contacts.keys():
    print (i)
```

```python
for k, v in contacts.items():
    print (k +' : '+v)
```

```python
#dicts comprehension
international = {k: "+1-"+ v for k,v in contacts.items() }
```

```python
for k, v in international.items():
    print( k +' number is '+v)
```

```python
# create a dictionary with keys the number from 0 to 100, 
# and values the squares of those numbers

dicts = {k: k**2 for k in range(0,100)}
print(dicts)
```
