```python
# Most bottles are 10 to 25 years old.
sns.distplot(data['age'], kde=False, color='orange', bins=10)
plt.title('Age distribution', fontsize=14)
plt.xlabel('Age-->', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.axvline((data.age).mean(), color="k", linestyle="--");
```

```python
#Alcohol percent or ABV distribtion
#
sns.distplot(data['Abv'], kde=False, color='orange', bins=10)
plt.title('ABV% Distribution', fontsize=14)
plt.xlabel('Alcohol percentage', fontsize=12)
plt.ylabel('Distribution', fontsize=12)
plt.axvline((data.Abv).mean(), color="k", linestyle="--");
```

```python
# Points distribution of the whiskies

# Shows no whisky is actually bad just that one tastes better than the others 

sns.distplot(data['points'], kde=False, color='brown', bins=10)
plt.title('Average score out of 100', fontsize=14)
plt.xlabel('Points count', fontsize=12)
plt.ylabel('Total bottles with score', fontsize=12)
plt.axvline((data.points).mean(), color="k", linestyle="--");
```

```python
# Price distribution  of the whiskies with KDE

# A histogram aims to approximate the underlying probability density function that generated the data by binning and counting observations. Kernel density estimation (KDE) presents a different solution to the same problem. Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernel, producing a continuous density estimate:

sns.distplot(data['price'], kde=True, color='green', bins=1, kde_kws={"color": "k", "lw": 3, "label": "KDE"})
plt.title('Average price(Dollars)', fontsize=14)
plt.xlabel('Price--->', fontsize=12)
plt.ylabel('Density', fontsize=12)

```

Check the description column to find the most common ingredients/flavours/tastes added to a single malt and blended whisky of a scotch whisky sin

Source code Inspiration :  https://www.datacamp.com/community/tutorials/wordcloud-python

```python
import sys
#!{sys.executable} -m pip install wordcloud

from wordcloud import WordCloud, STOPWORDS

def create_word_cloud(df, bg_color, max_words, mask, stop_words, max_font_size, colormap):
    
    wc = WordCloud(background_color = bg_color, max_words = max_words, mask = mask, stopwords = stop_words, max_font_size = max_font_size)
    wc.generate(' '.join(df))
    
    return wc.recolor(colormap = colormap, random_state = 42)
```

```python
sm_df = data[data['category'] == 'Single Malt Scotch']['description'].values
bd_df = data[data['category'] == 'Blended Scotch Whisky']['description'].values
```

```python
from PIL import Image
import requests # https://stackoverflow.com/questions/12020657/how-do-i-open-an-image-from-the-internet-in-pil/12020860

mask_sm = np.array(Image.open(requests.get('https://imgur.com/dAYKIVT.png', stream = True).raw))  # mask for single malts
mask_bd = np.array(Image.open(requests.get('https://imgur.com/upL1TBW.png', stream = True).raw))  # mask for blended whiskies
```

```python
stop_words = list(STOPWORDS)

fig, ax = plt.subplots(2, 1, figsize = (7, 14))

ax[0].imshow(create_word_cloud(df = sm_df, bg_color = 'white', max_words = 50, mask = mask_sm, stop_words = stop_words,
                             max_font_size = 50, colormap = 'winter'), alpha = 1, interpolation = 'bilinear')
ax[0].set_title('Single Malts - Initial', size = 16, y = 1.06)
ax[0].axis('off')

ax[1].imshow(create_word_cloud(df = bd_df, bg_color = 'white', max_words = 50, mask = mask_bd, stop_words = stop_words,
                             max_font_size = 50, colormap = 'tab10'), alpha = 1, interpolation = 'bilinear')
ax[1].set_title('Blended Whiskies - Initial', size = 16, y = 1.04)
ax[1].axis('off');
```

Words such as whisky, palate, note etc. aren't really informative. Adding them to stopwords and re-generating the cloud:

```python
stop_words = ['whisky', 'whiskies', 'blend', 'note', 'notes', 'year', 'years', 'old', 'nose', 'finish', 'bottle',
              'bottles', 'bottled', 'along', 'release', 'flavor', 'cask', 'well', 'make', 'mouth', 'palate', 'hint',
              'one', 'bottling', 'distillery', 'quite', 'time', 'date', 'show', 'first'] + list(STOPWORDS)
```

```python
fig, ax = plt.subplots(2, 1, figsize = (7, 14))

ax[0].imshow(create_word_cloud(df = sm_df, bg_color = 'white', max_words = 50, mask = mask_sm, stop_words = stop_words,
                             max_font_size = 50, colormap = 'winter'), alpha = 1, interpolation = 'bilinear')
ax[0].set_title('Single Malts - Improved', size = 16, y = 1.06)
ax[0].axis('off')

ax[1].imshow(create_word_cloud(df = bd_df, bg_color = 'white', max_words = 50, mask = mask_bd, stop_words = stop_words,
                             max_font_size = 50, colormap = 'tab10'), alpha = 1, interpolation = 'bilinear')
ax[1].set_title('Blended Whiskies - Improved', size = 16, y = 1.04)
ax[1].axis('off');
```

Observations:- 
    
The most common flavours/ingredients/tastes added to the Single Malt Scotches are vanilla, fruits, spices, ginger, chocolate

while for Blended whiskies are:  orange, oak, honey, caramel, chocolate. 

Differences: Sublte, floral, matured, soft

```python
A One sample t-test : testing the mean of a single group against a known mean.
```

value of Î± is 0.05. 

CI :  95% confidence that the conclusion of this test will be valid.

Null Hypothesis (H0) : Average Alcohol percentage (ABV) is less than or equal to 43% for all scotch whiskies.

H1 : Average  Alcohol percentage is not less than 43% for all scotch whiskies


```python
from scipy import stats
from scipy.stats import ttest_1samp
tstats, pvalue = stats.ttest_1samp(data.Abv, 43)
print(tstats, pvalue)
data.Abv.mean()
```

Result: P is less than 0.05 so we can reject H0.
