```python
import pandas as pd
import numpy as np
import os
import glob
```

```python
directory = 'C:/Users/svvan/Desktop/ProjectW6/Data1/Stocks'
files = []
for filename in os.listdir(directory):
    files.append(filename)

strings = ["a", "ab", "aa", "c"]

files2 = []

for filename in files:
    newname = filename.replace(".txt", "")
    files2.append(newname)
```

```python
path = 'C:/Users/svvan/Desktop/ProjectW6/Data1/Stocks'

# array = []
# for filename in files2:
#     data = pd.read_csv(filename, sep=" ", header=0)
#     array.append(data)

    
all_files = glob.glob(os.path.join(path,"*.txt"))
array = []

for filename in all_files:
    data = pd.read_csv(filename, sep=",", index_col=None, header=0)
    data['Name'] = filename
    array.append(data)
    
#     print(filename)

# data.columns = ["Date","Open","High","Low","Close","Volume","OpenInt"]
```

```python
merge_values = np.vstack(array)
data = pd.DataFrame(merge_values)
data.columns = ["Date","Open","High","Low","Close","Volume","OpenInt","Name"]
data['Name'] = data['Name'].str.replace("C:/Users/svvan/Desktop/ProjectW6/Data1/Stocks","")
data['Name'] = data['Name'].str.replace(".txt","")
```
