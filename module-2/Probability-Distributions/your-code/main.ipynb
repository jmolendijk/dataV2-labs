A [discrete random variable](https://en.wikipedia.org/wiki/Random_variable#Discrete_random_variable) has random numeric values that are not continuous. For example, the number of people in a household can only be integers but not floats. Therefore the number of people in a household is a discrete variable.

# Challenge 4 - Central Limit Theorem

A delivery company needs 35 minutes to deliver a package, with a standard deviation of 8 minutes. Suppose that in one day, they deliver 200 packages.

**Hint**: `stats.norm.cdf` can help you find the answers.

#### Step 1: What is the probability that the mean delivery time today is between 30 and 35 minutes?

```python

prob_30 = norm.cdf(30, loc=35, scale=8)
prob_35 = norm.cdf(35, loc=35, scale=1)

answer = prob_35 - prob_30
answer
```

#### Step 2: What is the probability that in total, it takes more than 115 hours to deliver all 200 packages?

```python
#Because everything above is in minutes we need to convert the 115 hours to minutes. 

minutes = 115*60
time_per_pack = minutes/200
#34,50 minuten per pack

prob_pack = norm.cdf(time_per_pack, loc=35, scale=8)

#Takes more than 115 --> is looking to the right of our x-axis (just a thing to remember)

total = 1-prob_pack
total
```
