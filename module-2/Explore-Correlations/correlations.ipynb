```python
# 2. Print a heat map of this correlation matrix (hint: use seaborn.heatmap())
sns.heatmap(cdd_df.corr())
```

```python
# 3. Find 3 pairs of columns, one pair that is highly positively correlated, one for which the correlation is almost 
# zero and one pair that is negatively correlated. In each case write down an explanation why we see this correlation



cdd_df['FATALITIES'].corr(cdd_df['INJURED / INFECTED'])
plt.scatter(x=cdd_df['FATALITIES'],y=cdd_df['INJURED / INFECTED'])

"""Fatalities and count of injured are strongly correlated. 
For a bigger disaster, both fatalities and count of injured will be high, and for a smaller disater they will be low"""
```

```python

cdd_df['FATALITIES'].corr(cdd_df['UTILITY - PEOPLE AFFECTED'])
plt.scatter(x=cdd_df['FATALITIES'],y=cdd_df['UTILITY - PEOPLE AFFECTED'])

"""Weak or no correlation between fatalities and count of people whose utilities are affected. 
Interruptions to electricity and water services are not related to the fatalites of a disaster"""
```

```python
cdd_df['OGD COSTS'].corr(cdd_df['UTILITY - PEOPLE AFFECTED'])
plt.scatter(x=cdd_df['OGD COSTS'],y=cdd_df['UTILITY - PEOPLE AFFECTED'])

"""Weak negative correlation between OGD costs and count of people whose utilities are affected. 
Interruptions to electricity and water services could be reduced when governments increase spending on the disater management"""
```
