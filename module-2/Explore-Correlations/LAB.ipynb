```python
#Find 3 pairs of columns, one pair that is highly positively correlated, 
#one for which the correlation is almost zero and one pair that is negatively correlated. 
#In each case write down an explanation why we see this correlation

#Highly positively correlated pair:

cdd_high = cdd[['INJURED / INFECTED', 'FATALITIES']]
cdd_high

"""
This one is highly correlated because in most of the cases recorded, when there is no injuries, there is no fatalities.
"""

#Correlation close to zero:

cdd_zero = cdd[['INJURED / INFECTED', 'ESTIMATED TOTAL COST']]
cdd_zero.head(60)

"""
The number of injured or infected persons has no meaningful impact on the total cost of a disaster.
"""

#Negative Correlation:

cdd_neg = cdd[cdd['UTILITY - PEOPLE AFFECTED'] >=0]
cdd_neg = cdd_neg[['UTILITY - PEOPLE AFFECTED', 'OGD COSTS']]
cdd_neg.head()

"""
This pair has a negative correlation, but it is so close to zero that looking at the data doesn't give 
us any insight about the link between the number of people whose utility had been affected, and the costs for
'Other Government Dapartment'.
"""
```

```python
#4. Look up the pairplot function in Seaborn. 
#Use it to plot the pairwise relations between the six columns you selected. 
#Does what you see confirm your ideas? Describe in a comment.

sns.pairplot(cdd_high)
#We see here the correlation : a high number or fatalities induced a high number of injured and vice versa.

sns.pairplot(cdd_zero)
#Here, the values have no 'organisation' on the graph below, they do not follow a 'trend'.

sns.pairplot(cdd_neg)
#Again here, we can't really see the negative correlation, as the values don't have any and are scattered. 
#Which confirm the fact that it is better to consider this pair as not correlated than negatively correlated. 
```
