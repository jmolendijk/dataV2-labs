### Looking at the distributions of the data
Let's take a look at the distribution of the 4 stats that describe the physical characteristics of the players.

**Plot the four distributions about `height`, `weight`, `age` and `BMI`.**

```python
sns.pairplot(wnba,vars=('Height','Weight','Age','BMI'))
```

```python
#your code here
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

fig,axs =plt.subplots(2,2, figsize=(8,10))

axs[0,0].hist(wnba['Height'],color='r',bins =10)#top left graph
plt.sca(axs[0,0])
plt.title("Player's Heights")
plt.xlabel('Height in cm')
plt.xticks(np.arange(160,205,10))
plt.ylabel('Count')

axs[1,0].hist(wnba['Weight'],color='y',bins=10)
plt.sca(axs[1,0])
plt.title("Player's Weight")
plt.xlabel('Weight in kg')
plt.xticks(np.arange(50,120,5))
plt.ylabel('Count')

axs[0,1].hist(wnba['Age'],color='b',bins=20)
plt.sca(axs[0,1])
plt.title("Player's Age")
plt.xlabel('Age range')
plt.xticks(np.arange(20,40,2))
plt.ylabel('Count')

axs[1,1].hist(wnba['BMI'], bins=13)
plt.sca(axs[1,1])
plt.title("Player's Body Mass")
plt.xlabel('BMI range')
plt.ylabel('Count')


plt.show()
```

**What conclusions do you think we can take from this plots?**

```python
#your conclusions here
'''The graphs show us standard distributions for all 4 characteristics.
On the funny note:
Most players measure between 180-190 (very few measuring higher than 2m
most players are between 22 and 28 years old)
Most players between 75 and 90kg
BMI between 18 and 25 is normal 25 to 30 is overweight (for the average 
height 186,25years, weight of 86, so taken all these average measures,
the average player is in the limit to be overweight! haha)
'''
```

In addition to what the describe() method already told us we can see that the physical characteristics of the players (apart from the age) more or less follow a standard distribution, which is expected when looking at the distribution of values that arise from a lot of factors that are independent from each other as is the case for many physical quantities. 

The height distribution looks like a bimodal. This may be due to the fact that players in basketball fall under two main categories (please note that this is a very gross generalization): shorter and more agile, and taller and less agile. Therefore there are less "average" height players since they will neither be as agile as the shorter players nor have the same impact in the paint (that is, under the basket) as a taller player.

The age distribution is a bit skewed to the right which is expected since most professional players stop playing after their prime physical years come to an end.

We could do the same for the main game stats. They are points, assists, blocks, rebounds and steals. 

**Now plot the distribution of the columns `REB`, `AST`, `STL`, `PTS` and `BLK` the same way you did in the last cell.**

```python
#your code here

fig,axs =plt.subplots(2,3, figsize=(12,10))

axs[0,0].hist(wnba['REB'],color='b',bins=20)
plt.sca(axs[0,0])
plt.title("Player's Rebounds")
plt.xlabel('rebounds')
plt.ylabel('Count')

axs[1,0].hist(wnba['AST'],color='m',bins=10)
plt.sca(axs[1,0])
plt.title("Player's Assists")
plt.xlabel('Assists')
plt.ylabel('Count')

axs[0,1].hist(wnba['PTS'],color='y',bins=20)
plt.sca(axs[0,1])
plt.title("Player's Points")
plt.xlabel('points')
plt.ylabel('Count')

axs[1,1].hist(wnba['BLK'],color='c', bins=13)
plt.sca(axs[1,1])
plt.title("Player's Blocks")
plt.xlabel('Blocks')
plt.ylabel('Count')

axs[0,2].hist(wnba['STL'], bins=13)
plt.sca(axs[0,2])
plt.title("Player's Steals")
plt.xlabel('Steals')
plt.ylabel('Count')


```

**What conclusions do you think we can take from this plots?**

As expected all of the above distribution are heavily skewed to the right, since most players will have very low to average stats while there will be a handful of star players whose stats peak above everyone else. It is also important to think about the fact that we are simply taking the stats as they are without considering the minutes played by each player. Even though skill plays a very important factor in determining this kind of stats we also have to consider that players that play more minutes will, on average, score more points (or blocks, assists, etc.).

**For the sake of it let's look at the same distributions by dividing those stats by the minutes played for each player in the dataset.** 

```python
#your code here
# REB, AST, STL, PTS and BLK

reb2 = wnba['REB']/wnba['MIN']
ast2 = wnba['AST']/wnba['MIN']
stl2 = wnba['STL']/wnba['MIN']
pts2 = wnba['PTS']/wnba['MIN']
blk2 = wnba['BLK']/wnba['MIN']

fig,axs =plt.subplots(2,3, figsize=(12,10))

axs[0,0].hist(reb2,color='b',bins=20)
plt.sca(axs[0,0])
plt.title("Player's Rebounds by minutes played")
plt.xlabel('rebounds')
plt.ylabel('Count')

axs[1,0].hist(ast2,color='m',bins=10)
plt.sca(axs[1,0])
plt.title("Player's Assists by minutes played")
plt.xlabel('Assists')
plt.ylabel('Count')

axs[0,1].hist(pts2,color='y',bins=20)
plt.sca(axs[0,1])
plt.title("Player's Points by minutes played")
plt.xlabel('points')
plt.ylabel('Count')

axs[1,1].hist(blk2,color='c', bins=13)
plt.sca(axs[1,1])
plt.title("Player's Blocks by minutes played")
plt.xlabel('Blocks')
plt.ylabel('Count')

axs[0,2].hist(stl2, bins=13)
plt.sca(axs[0,2])
plt.title("Player's Steals by minutes played")
plt.xlabel('Steals')
plt.ylabel('Count')


```
