#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
y2 = x**2
ye = np.exp(x)

fig, ax2 = plt.subplots()

ax2.plot(x, y2, color='green')
ax2.set_xlabel('X')
ax2.set_ylabel('SQUARE', color='green')
ax2.set_yscale('linear')
ax2.tick_params(axis='y', labelcolor='green')

axe = ax2.twinx()

axe.plot(x, ye, color='blue')
axe.set_xlabel('X')
axe.set_ylabel('EXP', color='blue')
axe.set_yscale('log')
axe.tick_params(axis='y', labelcolor='blue')

#fig.tight_layout()
plt.show()
```

#### As a bonus challenge, try to add a legend to the plot.

```python
y2 = x**2
ye = np.exp(x)

fig, ax2 = plt.subplots()

ax2.plot(x, y2, color='green', label='Square')
ax2.set_xlabel('X')
ax2.set_ylabel('SQUARE', color='green')
ax2.set_yscale('linear')
ax2.tick_params(axis='y', labelcolor='green')
plt.legend()
axe = ax2.twinx()

axe.plot(x, ye, color='blue', label='Exp')
axe.set_xlabel('X')
axe.set_ylabel('EXP', color='blue')
axe.set_yscale('log')
axe.tick_params(axis='y', labelcolor='blue')

#fig.tight_layout()
plt.legend() #got both legends at the same place
plt.show()
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# MEDIAN STEPS BY MONTH_ENCODED
med_steps = fitbit.pivot_table(index=['Months_encoded'], values=['Steps'], aggfunc='median')
plt.figure(figsize=(12,4))
plt.plot(med_steps, label='MEDIAN STEPS BY MONTH_ENCODED')
plt.xlabel('Months')
plt.ylabel('Median Steps')
plt.legend()
plt.savefig('MEDIAN STEPS BY MONTH_ENCODED')
plt.show()


# MEDIAN STEPS BY WORK_OR_WEEKEND
med_steps_wk = fitbit.pivot_table(index=['Work_or_Weekend'], values=['Steps'], aggfunc='median')
plt.figure(figsize=(12,4))
plt.plot(med_steps_wk, label='MEDIAN STEPS BY WORK_OR_WEEKEND')
plt.xlabel('Work or Weekend')
plt.ylabel('Median Steps')
plt.legend()
plt.savefig('MEDIAN STEPS BY WORK_OR_WEEKEND')
plt.show()

```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
minutes = ['Minutes Lightly Active', 'Minutes Very Active', 'Minutes Sedentary']

for m in minutes:
    plt.figure(figsize=(12,4))
    plt.scatter(fitbit['Steps'], fitbit[m], label=m+' vs Steps')
    plt.xlabel('Steps')
    plt.ylabel(m)
    plt.legend()
    plt.show()
```
