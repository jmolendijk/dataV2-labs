```python
steps_stat_1 = fitbit[['Steps', 'Months_encoded']].groupby('Months_encoded').median()
steps_stat_1
```

```python
steps_stat_2 = fitbit[['Steps', 'Work_or_Weekend']].groupby('Work_or_Weekend').median()
steps_stat_2
```

```python
# These plots are made using 

x1 = steps_stat_1.reset_index()['Months_encoded'].to_numpy()
y1 = steps_stat_1.reset_index()['Steps'].to_numpy()

x2 = steps_stat_2.reset_index()['Work_or_Weekend'].to_numpy()
y2 = steps_stat_2.reset_index()['Steps'].to_numpy()

f, ax = plt.subplots(2,1, figsize=(8,4))  # I didn't really like (12,4) size, I think this looks more organic.

plt.sca(ax[0])
plt.plot(x1,y1)
plt.sca(ax[1])
plt.xticks(np.arange(2), ('Weekend', 'Working day'))
plt.bar(x2,y2)
plt.show()
```

```python
# Another way to solve it, it some ways easier, in others - more fiddly.

# MEDIAN STEPS BY MONTH_ENCODED
# your code here
steps_stat_1 = fitbit[['Steps', 'Months_encoded']].groupby('Months_encoded').median()

# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here
steps_stat_2 = fitbit[['Steps', 'Work_or_Weekend']].groupby('Work_or_Weekend').median()
steps_stat_2.rename(index={0: 'Weekend', 1: 'Working Day'}, inplace=True)

figure, axes = plt.subplots(1, 2, figsize=(12,4))
steps_stat_1.plot(ax=axes[0])

axes = steps_stat_2.plot(kind='bar', ax=axes[1])
axes.legend(['Steps'])
```
