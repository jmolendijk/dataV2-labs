#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

#### As a bonus challenge, try to add a legend to the plot.

```python
# your code here
x = np.arange(0,100)
y = x**2
z = math.e ** x

fig, [xy, xz] = plt.subplots(1,2, figsize = (12, 6))
xy.plot(x,y)
xy.set_title('Square') 
xy.set_xlabel('Label x')
xy.set_ylabel('Label y')
xy.set_yscale('linear')

xz.plot(x,z)
xz.set_title('Log') 
xz.set_xlabel('Label x')
xz.set_ylabel('Label y')
xz.set_yscale('log')

```

```python
# your code here
#

x = np.arange(0,100)
y = x**2
z = math.e ** x

fig, [xy, xz] = plt.subplots(1,2, figsize = (12, 6))
l1, = xy.plot(x,y)
xy.set_title('Square') 
xy.set_xlabel('Label x')
xy.set_ylabel('Label y')
xy.set_yscale('linear')


l2, = xz.plot(x,z)
xz.set_title('Log') 
xz.set_xlabel('Label x')
xz.set_ylabel('Label y')
xz.set_yscale('log')

xy.legend((l1, l2), ('square', 'log'), loc='upper left')
```
