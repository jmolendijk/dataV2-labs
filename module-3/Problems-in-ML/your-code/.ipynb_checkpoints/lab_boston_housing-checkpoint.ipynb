### Import and explore the data. Clean the data for outliers and missing values. 

```python
# Your code here
import pandas as pd
import numpy as np

data = pd.read_csv('../data/boston_data.csv')
data.head()
```

```python
data.info() 
# No null values and all float type
# chas was a categorical column converted to numeric using dummy values all the other columns are non-categorical.
```

```python
np.where(data.applymap(lambda x: x == '')) # No empty values
```

```python
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1
print(IQR)

# Here we have the interquartile range for each column. We need this so we can look for outliers. We assume a datapoint to be
# an outlier when it is either smaller than Q1-1.5*IQR or bigger than Q3+1.5*IQR.
```

```python
print(data < (Q1 - 1.5 * IQR)) |(data > (Q3 + 1.5 * IQR))
# False = no outlier
# True = outlier
```

```python
data_new = data[~((data < (Q1 - 1.5 * IQR)) |(data > (Q3 + 1.5 * IQR))).any(axis=1)]
data_new
# Here we defined a new DF where only the values are kept when they are bigger than Q1-1.5*IQR or smaller than Q3+1.5*IQR.
```

```python
# Your code here
from sklearn.model_selection import train_test_split

X = data_new.drop(columns=['medv'])
y = data_new['medv']

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, shuffle=True)
```

## Analyzing Model Performance
Next, we are going to build a Random Forest Regressor, and test its performance with several different parameter settings.

### Learning Curves
Lets build the different models. Set the max_depth parameter to 2, 4, 6, 8, and 10 respectively. 

```python
# Five separate RFR here with the given max depths
from sklearn.ensemble import RandomForestRegressor

modeld2 = RandomForestRegressor(max_depth=2)
modeld4 = RandomForestRegressor(max_depth=4)
modeld6 = RandomForestRegressor(max_depth=6)
modeld8 = RandomForestRegressor(max_depth=8)
modeld10 = RandomForestRegressor(max_depth=10)

modeld2.fit(X_train, y_train)
modeld4.fit(X_train, y_train)
modeld6.fit(X_train, y_train)
modeld8.fit(X_train, y_train)
modeld10.fit(X_train, y_train)
```

Now, plot the score for each tree on the training set and on the testing set.

```python
# Produce a plot with the score for the testing and training for the different max depths
models = ['modeld2','modeld4','modeld6','modeld8','modeld10']
modeld2_scores = []
modeld4_scores = []
modeld6_scores = []
modeld8_scores = []
modeld10_scores = []
scores = [modeld2_scores,modeld4_scores,modeld6_scores,modeld8_scores,modeld10_scores]


y_pred2train = modeld2.predict(X_train)
modeld2_scores.append(r2_score(y_train, y_pred2train))
y_pred2test = modeld2.predict(X_test)
modeld2_scores.append(r2_score(y_test, y_pred2test))

y_pred4train = modeld4.predict(X_train)
modeld4_scores.append(r2_score(y_train, y_pred4train))
y_pred4test = modeld4.predict(X_test)
modeld4_scores.append(r2_score(y_test, y_pred4test))

y_pred6train = modeld6.predict(X_train)
modeld6_scores.append(r2_score(y_train, y_pred6train))
y_pred6test = modeld6.predict(X_test)
modeld6_scores.append(r2_score(y_test, y_pred6test))

y_pred8train = modeld8.predict(X_train)
modeld8_scores.append(r2_score(y_train, y_pred8train))
y_pred8test = modeld8.predict(X_test)
modeld8_scores.append(r2_score(y_test, y_pred8test))

y_pred10train = modeld10.predict(X_train)
modeld10_scores.append(r2_score(y_train, y_pred10train))
y_pred10test = modeld10.predict(X_test)
modeld10_scores.append(r2_score(y_test, y_pred10test))
```

```python
df = pd.DataFrame(columns=models)
df['modeld2'] = modeld2_scores
df['modeld4'] = modeld4_scores
df['modeld6'] = modeld6_scores
df['modeld8'] = modeld8_scores
df['modeld10'] = modeld10_scores
df = df.transpose()
df = df.rename(columns={0:'Training_Score',1:'Test_Score'})
df.transpose()
df.reset_index()
```
